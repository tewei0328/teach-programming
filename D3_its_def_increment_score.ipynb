```python
def increment_score(score, bonus, points):
  if bonus == True:
    points = points * 2
    score = score + points
  return score
```

```python
points = 5
```

```python
score = 10
```

```python
increment_score(score, True, points)
```

```python
new_score = increment_score(score, True, points)
```

```python
print(new_score)
```
