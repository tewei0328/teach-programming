```python
import google.generativeai as genai
from docx import Document
import io
import PIL.Image

# 1. 配置 API Key (請替換為你的 API Key)
genai.configure(api_key="...")

def image_to_word(image_path, output_path):
    # 2. 初始化模型 (使用支援圖像的 Flash 模型)
    model = genai.GenerativeModel('gemini-2.5-flash-image')

    # 3. 載入手寫照片
    img = PIL.Image.open(image_path)

    # 4. 定義 Prompt：要求 AI 辨識並以 Markdown 表格輸出
    prompt = """
    請辨識這張照片中的手寫表格內容。
    1. 請精確提取日期、時間、收縮壓、舒張壓、脈搏等數據。
    2. 請直接以 Markdown 表格格式輸出，不要有額外的解釋文字。
    3. 如果字跡潦草，請根據上下文邏輯推斷最合理的數字。
    """

    # 5. 呼叫 API
    response = model.generate_content([prompt, img])
    table_text = response.text
    print("--- 辨識結果 ---")
    print(table_text)

    # 6. 將 Markdown 轉存為 Word 檔
    doc = Document()
    doc.add_heading('血壓紀錄自動辨識', 0)

    # 解析 Markdown 表格並填入 Word (簡單處理邏輯)
    lines = [line.strip() for line in table_text.split('\n') if '|' in line]
    if lines:
        # 取得標題與數據列
        rows_data = []
        for line in lines:
            if '---' in line: continue # 跳過 Markdown 的分割線
            cells = [c.strip() for c in line.split('|') if c.strip()]
            rows_data.append(cells)

        if rows_data:
            table = doc.add_table(rows=len(rows_data), cols=len(rows_data[0]))
            table.style = 'Table Grid'
            for i, row in enumerate(rows_data):
                for j, cell_text in enumerate(row):
                    table.cell(i, j).text = cell_text

    doc.save(output_path)
    print(f"\n文件已成功儲存至: {output_path}")

# 執行
image_to_word('111.jpg', 'blood_pressure_record.docx')

# 移除列出可用模型的步驟，因為已經找到並使用了合適的模型
```
