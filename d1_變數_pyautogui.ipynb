```python
from pyvirtualdisplay import Display
import os

# Create a dummy .Xauthority file to prevent XauthError
# This is a common workaround in headless environments like Colab
authority_path = os.path.expanduser('~/.Xauthority')
if not os.path.exists(authority_path):
    open(authority_path, 'a').close()
    os.chmod(authority_path, 0o600) # Set appropriate permissions

# 啟動虛擬顯示器
vdisplay = Display(visible=0, size=(1920, 1080))
vdisplay.start()

import pyautogui
# 現在應該可以成功 import 了
```
